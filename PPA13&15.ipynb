```python
def subset_sum(L, s, test='Not Used'):

  print('DEBUG 🐛⚔️', L, s, test)

  if s == 0:
    return True
  if len(L) == 0:
    return False
  return subset_sum(L[1:], s - L[0], test=f'Used {L[0]}') or subset_sum(L[1:], s)
```

```python
def power(matrix, exponent):
  if exponent == 0:
    size = len(matrix)
    return [[1 if i == j else 0 for j in range(size)] for i in range(size)]
  else:
    return matrix_multiply(matrix, power(matrix, exponent - 1))


def matrix_multiply(A, B):
  result = []
  for i in range(len(A)):
    new_row = []
    for j in range(len(B[0])):
      new_element = sum(A[i][k] * B[k][j] for k in range(len(B)))
      new_row.append(new_element)
    result.append(new_row)
  return result
```

```python
# Example matrix A
A = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
]

# Example usage of matrix multiplication
matrix_multiply(A, A)  # Multiply A by itself
matrix_multiply(A, matrix_multiply(A, A))  # Multiply A by A^2
```

```python
# Example usage of matrix power calculation
power(A, 1)  # A^1
power(A, 2)  # A^2
power(A, 3)  # A^3
power(A, 4)  # A^4
```
